## **Reglas de asociacin? Qu es eso!**

>**El aprendizaje de reglas de asociacin es un mtodo de aprendizaje automtico basado en reglas para descubrir relaciones interesantes entre variables en grandes bases de datos. Su objetivo es identificar reglas slidas descubiertas en bases de datos utilizando algunas medidas de inters.**

Basndose en el concepto de reglas estrictas, **Rakesh Agrawal, Tomasz Imieliski y Arun Swam** introdujeron reglas de asociacin para descubrir regularidades entre productos en datos de transacciones a gran escala registrados por sistemas de punto de venta (POS) en los supermercados. Por ejemplo, la regla **{cebollas, patatas} -> {hamburguesa}** que se encuentra en los datos de ventas de un supermercado indicara que si un cliente compra cebollas y patatas juntas, es probable que tambin compre carne de hamburguesa. Dicha informacin se puede utilizar como base para decisiones sobre actividades de marketing como, **por ejemplo, precios promocionales o colocacin de productos**.

Adems del ejemplo anterior del anlisis de la canasta de mercado, las reglas de asociacin se emplean hoy en muchas reas de aplicacin, incluida la minera del uso de la web, la deteccin de intrusiones, la produccin continua y la bioinformtica. A diferencia de la minera de secuencias, el aprendizaje de reglas de asociacin generalmente no considera el orden de los elementos dentro de una transaccin o entre transacciones. [Wikipedia]

## **MBA (Market Basket Analysis):**
> **El anlisis de la canasta de mercado examina los productos que los clientes tienden a comprar juntos y utiliza la informacin para decidir qu productos se deben vender o promocionar juntos..)**

**Reglas de asociacin**

Si bien estamos interesados en extraer conjuntos frecuentes de elementos, esta informacin a menudo se presenta como una coleccin de reglas **si-entonces**, llamadas **reglas de asociacin**.

La forma de una regla de asociacin es $\{X\rightarrow Y\}$, donde $\{X\}$ es un conjunto de elementos y $\{Y\}$ es un elemento. La implicacin de esta regla de asociacin es que si todos los elementos de $\{X\}$ aparecen en alguna canasta, entonces $\{Y\}$ es probable que tambin aparezcan en esa canasta.

*	$\{X\}$ tambin se llama antecedente o lado izquierdo (LHS) y

* $\{Y\}$ se llama consecuente o lado derecho (RHS) .

Un ejemplo de regla de asociacin para productos de Apple podra ser $\{Apple iPad, Apple,iPad, Cover\}\rightarrow \{Apple, Pencil\}$, lo que significa que si $\{X\}$ se compran la funda para iPad y iPad de Apple , es probable que los clientes tambin compren el lpiz de Apple $\{Y\}$. Observe que el smbolo de implicacin lgica 
$\rightarrow$ no indica una relacin causal entre $\{X\}$ e $\{Y\}$. Es simplemente una estimacin de la probabilidad condicional de $\{Y\}$ dado $\{X\}$.

**Medidas de asociacin**

Hay tres criterios ms populares que evalan la calidad o la fuerza de una regla de asociacin: apoyo, confianza y sustentacin.

***Support*** 
> Es el porcentaje de transacciones que contienen una determinada combinacin de elementos en relacin con el nmero total de transacciones en la base de datos.
Tambin nos dice qu tan popular es un conjunto de elementos, medido por la proporcin de transacciones en las que aparece el conjunto de elementos.
O en general, para un conjunto de elementos $X$:

$$support(X)=\frac{Frecuencia(X)}{n} $$

>con $n =$ nmero de todas las transacciones (cestas).

>Por lo general, se utiliza un umbral de soporte especfico para reducir la cantidad de conjuntos de elementos que necesitamos analizar. Al comienzo del anlisis, podramos establecer nuestro umbral de soporte al 10%.

***Confidence*** 
$$X \to Y$$
> Mide cunto depende el consecuente (elemento) del antecedente (elemento). Es la probabilidad condicional del consecuente dado el antecedente, $P(B|A)$. Por ejemplo, la confianza para la regla de asociacin si pan entonces helado es del 66% ya que tres transacciones contienen pan (el antecedente) y dos de las tres transacciones tambin contienen helado (el consecuente). Es decir, dado que se seleccionan las canastas que contienen helado, existe un 66% de posibilidades de que la misma canasta tambin contenga helado.

>La confianza nos dice la probabilidad de que se compre un artculo $Y$ dado que se compra el artculo $X$, expresado como $\{X\rightarrow Y\}$. Se mide por la proporcin de transacciones con el artculo $X$, en el que tambin aparece el artculo $Y$. La confianza de una regla se define como:

$$\ support(X\rightarrow Y)=\frac{support(X\cup Y)}{support(X)} $$

>Por tanto, la confianza se puede interpretar como una estimacin de la probabilidad $P(Y/X)$. En otras palabras, esta es la probabilidad de encontrar el RHS (Y) de la regla en transacciones con la condicin de que estas transacciones tambin contengan el LHS (X) (Hornik, Grn y Hahsler, 2005). La confianza est dirigida y por lo tanto suele dar valores diferentes a las reglas $X\rightarrow Y$ y $Y\rightarrow X$.
   

***Lift*** 
>(Tambin llamada mejora o impacto) es una medida para superar los problemas con apoyo y confianza. Se dice que Lift mide la diferencia, medida en proporcin, entre la confianza de una regla y la confianza esperada.

>la elevacin se calcula como:

$$Lift(XY)=\frac{support(X\cup Y)} {support(X) * support(Y)}$$

>**Reglas:**

     -Si la regla tuviera un lift de =1, entonces X y Y son independientes y no se puede derivar ninguna regla de ellos.
    
     - Si el lift es > 1, significa que es probable que se compre el artculo Y si se compra el artculo X, es decir,  P(Y/X) entonces A y B son dependientes entre s, y el grado de la cual viene dado por el valor de lift.
     
     - Si el lift es < 1, menos de 1 significa que es poco probable que se compre el artculo Y si se compra el artculo X, entonces la presencia de X tendr un efecto negativo en Y.

**Ejemplo:**

      Gorra => gafas de sol 
      * Support=20% ;
      * Confidence=60% ;  

     La regla anterior dice:
      La transaccin del 20% muestra que Sunglass se compra con la compra de un Gorra
      El 60 % de los clientes que compran gafas de sol lo compran con la compra de unas Gorra

*Cada criterio tiene sus ventajas y desventajas pero, en general, mantenemos reglas de asociacin que tienen un alto nivel de confianza, un alto apoyo y un alto nivel de sustentacin. Las reglas de asociacin con alto soporte son reglas potencialmente interesantes. De manera similar, las reglas con alta confianza seran reglas interesantes. O puede buscar reglas de asociacin con elevacin muy alta o muy baja.*

### 4. *Market Basket Analysis* : <a id="MBA"></a>

```python
basket.head()
```

```python
basket.shape
```

```python
df = basket.groupby(['Transaction','Item']).size().reset_index(name='count')
basket0 = (df.groupby(['Transaction', 'Item'])['count']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Transaction'))
basket0
```

```python
#The encoding function
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket0.applymap(encode_units)
```

Aqu hay dos funciones principales.
- La funcin `apriori ()` evala el valor de soporte para cada producto.
- La funcin `association_rules ()` nos ayuda a comprender la relacin entre antecedentes y productos de consecuencias. Proporciona informacin notable sobre los productos.

Especialmente, las variables de "soporte antecedente", "soporte consecuente", "soporte", "confianza" y "elevacin" son muy importantes para tomar algunas decisiones comerciales.

### *Un Antecedente* <a id="one"></a>

```python
#  Aplicamos la funcin apriori
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)

# aplicar la  regla de asociacin
rules = association_rules(frequent_itemsets, metric="lift")
rules.sort_values('confidence', ascending = False, inplace = True)
print(rules.shape)
rules
```

**Regla:**

Toast ==> coffee

suporte=2.35%

confidence=70%

- ***Segn la tabla anterior, se espera que los clientes que compraron (Toast) compren (coffee) con una probabilidad del 70% (confianza). Las puntuaciones de elevacin(lift) y conviccin tambin respaldan esa hiptesis***
- **Como el lift=1.48 que es mayor que 1, significa que es probable que se compre el artculo Toast si se compra cafe**
- 
- ***Sera mejor mantenerlos cerca para aumentar las ventas!***

```python
# Visualizacin del color de distribucin de reglas mapeado por Lift
plt.figure(figsize=(10, 6))
plt.scatter(rules['support'], rules['confidence'], c=rules['lift'], alpha=0.9, cmap='BuPu');
plt.title('Rules distribution color mapped by lift');
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.colorbar();
```

```python
rules[(rules.support > 0.01) & (rules.confidence > 0.15)].sort_values("confidence", ascending = False)
```

```python
# Establezca la mtrica "lift" con un umbral mnimo = 1.2

frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)
rules2 = association_rules(frequent_itemsets, metric="lift",  min_threshold=1.2)
rules2.sort_values('confidence', ascending = False, inplace = True)
print(rules2.shape)
rules2
```

```python
# Visualizacin del color de distribucin de reglas mapeado por Lift
plt.figure(figsize=(10, 6))
plt.scatter(rules2['support'], rules2['confidence'], c=rules2['lift'], alpha=0.9, cmap='BuPu');
plt.title('Rules distribution color mapped by lift');
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.colorbar();
```
