### Tipo: Outer

```python
# generamos data para nuestros ejemplos
df1 = df4.sample(4)
df1 = df1.reset_index(drop=True)   # 3 * 7
df1
```

```python
#import numpy as np
df2 = pd.DataFrame(
    np.random.randint(1,11, 25).reshape([5, 5]),
    columns = ["b", "c", "d", "e", "h"])

df2['ID'] = ["id1","id3", "id5", "id7", "id9"]
df2  # 3 * 4
```

Colocamos **sort = False** para evitar que se muestre un aviso al respecto de cierto cambio de funcionalidad que se produzaca en futuras versiones de Pandas.

```python
# Unimos los dos dataframes según la etiqueta de las columnas
pd.concat([df1, df2], sort= False)
```

```python
# Unimos los dos DataFrame's a lo largo de la columna
# Según el Index

pd.concat([df1, df2], axis = 1, sort = False)
```

```python
# Unimos los dos DataFrame's a lo largo de la columna
# Según 1 o varias columnas

df1.merge(df2, on=["ID"], how='outer')
```

### Tipo: Inner

```python
# Intersección (y) a nivel de columnas
pd.concat([df1, df2], join = "inner")  # Sólo las columnas que son comunes 
```

```python
# Intersección (y) a nivel de filas
pd.concat([df1, df2], axis = 1, join = "inner") #, ignore_index=True) # sólo las filas que son comunes
```

```python
# Unimos los dos DataFrame's a lo largo de la columna (i)
# Según 1 o varias columnas

df1.merge(df2, on=["ID"], how='inner')
```

### Tipo: Left

```python
# Unimos los dos DataFrame's: Base DataFrame Izquierdo
# Según 1 o varias columnas

df1.merge(df2, on=["ID"], how='left')
```

### Tipo: Right

```python
# Unimos los dos DataFrame's: Base DataFrame Derecho
# Según 1 o varias columnas

df1.merge(df2, on=["ID"], how='right')
```

# Pivot

Permite construir un nuevo dataframe a partir de otro, de tal forma que los valores de una columna pasan a ser nombres de columna del nuevo dataframe.

Fijar las características del nuevo DataFrame:
- Index 
- Columnas
- Valores

Muy útil para el tratamiento de datos.


```python
pd.pivot_table(df6, values="d", index=["f", "g"]) # 2 index cualitativas por la columna d
```

```python
# Forma 1
df6_pivot = pd.pivot_table(df6, values="d", index=["f", "g"], columns=["e"]) # 2 index por la columna d y como columna e
df6_pivot
```

# Datos Duplicados

El método duplicated de la clase DataFrame permite identificar filas repetidas; cuyo resultado es de tipo **Bool**, donde el valor **True** indica que ya existe, caso contrario será **False** .

datos duplicados.jpg

```python
df7 = pd.DataFrame([['Perú', 10200],  # generando data para nuestros ejemplos
       ['Brasil', 11506],
       ['Brasil', 11506],
       ['Chile', 5500],
       ['Perú', 10200],
       ['Chile', 23], 
       ['Ecuador', 10000],
       ['Perú', 2560]], 
       columns = ['Pais', 'Cifra'])
df7
```

```python
df7.duplicated() #Se está revisando a nivel de todas las columnas
```

```python
# Extraer en un nuevo DataFrame todos los duplicados
df7[df7.duplicated()] # True
```

```python
# Eliminar datos duplicados
df7.drop_duplicates() # nos deja el DataFrame sin los datos duplicados
```

```python
df7  # si se desea mantener el cambio, almacenar en un contenedor
```

```python
# Eliminar duplicados por ciertas columnas en específico

df7.drop_duplicates(subset = ['Pais'])  # mantuvo el 1ro y eliminó el resto de duplicados
```

```python
# Eliminar todas las filas repetidas, pero no la última "keep"

df7.drop_duplicates(subset = ['Pais'], keep = 'last') 
```
